# Exercício 02

Escreva um programa que receba como parâmetro de entrada um número inteiro de 
5 dígitos no intervalo fechado [10000, 30000] que represente códigos de produtos vendidos em uma loja. Crie uma função para validar os dados de entrada, obrigando o usuário a respeitar o intervalo e o tipo de dado (inteiro).

Crie uma função que calcule e retorne o dígito verificador do código, utilizando a regra de cálculo explicada a seguir. Por exemplo, considere o código 21853, em que cada dígito é multiplicado por um peso começando em 2, os valores obtidos são somados, e do total obtido calcula-se o resto de sua divisão por 7.

| Dígito        | 2 | 1 | 8  | 5  | 3  |                       |
|---------------|---|---|----|----|----|-----------------------|
| Peso          | 2 | 3 | 4  | 5  | 6  |                       |
| Multiplicação | 4 | 3 | 32 | 25 | 18 | Soma todos = 82       |
|               |   |   |    |    |    | Resto de 82 por 7 = 5 |


Retorne na função o valor do dígito verificador calculado e imprima na tela o código do produto digitado e seu dígito verificador separado por hífen, como: 21853-5.

Imprima na tela um teste do seu programa utilizando como código os 5 primeiros 
dígitos do seu RU. Se seu RU tiver menos de 5 dígitos, complete com zeros. Se seu RU cair fora do intervalo especificado, realize o teste mesmo assim.

```python
# Função para validação da entrada do usuário
# 
# Parâmetros:
#   - RU: flag para testar quando inserir o código do RU (padrão = False)
#       - Quando RU for True, desconsidera a verificação do limite inferior
#   - limiteInf: limite inferior do intervalo (padrão = 10000)
#   - limiteSup: limite inferior do intervalo (padrão = 30000)
#
def validaEntrada(RU=False, limiteInf=10000, limiteSup=30000):
    # Loop para receber e testar a entrada do usuário
    while True:
        try:
            num = int(input(f"Digite um número inteiro de 5 dígitos no intervalor fechado [{limiteInf}, {limiteSup}]: "))

            # Teste se a entrada é menor que o limite inferior
            if (num < limiteInf) and (not RU):
                print(f"O número deve ser maior ou igual a {limiteInf}.")

            # Teste se a entrada é maior que o limite superior
            elif (num > limiteSup) and (not RU):
                print(f"O número deve ser menor ou igual a {limiteSup}.")

            # Se a entrada for correta, interrompe o loop e retorna o número digitado
            else:
                break
        except:
            print("Digite um número inteiro.")

    return num

# Função para o cálculo do digito verificador
#
# Parâmetro: num: recebe o número digitado pelo usuário
#
def calculaDigito(num):
    # Inicialização das variáveis
    somaMult = 0
    peso = 2

    # Loop para multiplicar a somar cada número pelo seu respectivo peso
    for i in str(num):
        somaMult = somaMult + int(i)*peso
        peso += 1
    
    # Calcula o resto da soma das multiplicações por 7
    resto = somaMult % 7

    return resto

# Recebe e valida a entrada do usuário, RU=True irá ignorar os limites
numero = validaEntrada(RU=True)

# Completa com zeros, caso o número informado possua menos que 5 dígitos
numeroStr = str(numero).ljust(5, '0')

# Calcula o dígito verificador
digito = calculaDigito(numeroStr)

# Imprime o número informado e o dígito verificador
print(f"{numeroStr}-{digito}")
```
