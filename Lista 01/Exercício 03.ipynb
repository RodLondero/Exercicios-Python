# Exercício 03

Considere o seguinte conjunto de dados: Nome + (N1, N2, N3, N4). Nome representa o nome de um aluno e deve ser usado como chave. Já N1, N2, N3, N4 representam as notas de provas desse aluno. Utilize uma estrutura de dicionário com listas para resolver este exercício.

Escreva um programa que leia os dados de N alunos e apresente na tela se foram 
aprovados ou reprovados. O critério que garante a aprovação é que a média aritmética das 4 notas seja maior ou igual 7,0. O valor de N é a quantidade de alunos, e esse valor deve ser lido do teclado no começo do programa. Faça um laço de repetição para a leitura destes N alunos. As notas devem ser exibidas ao final do programa com uma casa decimal de precisão.


| Notas dos alunos   |     |     |     |     |     |           |
|--------------------|-----|-----|-----|-----|-----|-----------|
| Cláudio dos Santos | 6.5 | 3.5 | 5.0 | 6.5 | 5.4 | Reprovado |
| Eduardo Falco      | 7.5 | 6.5 | 7.0 | 6.5 | 6.9 | Aprovado  |
| Ligia de Oliveira  | 8.5 | 7.5 | 8.0 | 7.0 | 7.9 | Aprovado  |

Imprima na tela um teste do seu programa usando como primeiro cadastro o seu 
nome, e como nota os 4 primeiros dígitos do seu RU.


```python
# Loop para receber e validar a entrada do número de alunos
while True:
    try:
        n = int(input("Digite o número de alunos: "))
        break
    except:
        print("O número de alunos deve ser um número inteiro.")

# Inicialização das variáveis
alunos = list()
aluno = dict()

# Loop que percorre o número de alunos informados
for i in range(0, n):

    aluno['nome'] = input("\nNome do aluno(a): ")

    # Loop para percorrer a quantidade de notas (4)
    for nota in range(0, 4):
        # Loop para receber e validar a nota informada
        while True:
            try:
                # Formata a chave para receber a nota (n1, n2...)
                key_nota = f"n{nota+1}"

                aluno[key_nota] = float(input(f"Nota {nota+1}: "))
                
                # Verifica se a nota é menor que zero ou maior que 10
                if (aluno[key_nota] < 0) or (aluno[key_nota] > 10):
                    print("Digite uma nota entre 0 e 10.")
                else:
                    break
            except:
                print("Digite somente números.")

    # Inclui o dicionário aluno na lista
    alunos.append(aluno.copy())

# Imprime a lista dos alunos
print(f"\n{'Notas dos alunos':<20} {'N1':<6} {'N2':<6} {'N3':<6} {'N4':<6} {'Média':<10} {'Situação':<5}")
print("-"*60)

for aluno in alunos:
    soma = 0
    # Cálculo da média
    for nota in range(0, 4):
        soma = soma + aluno[f"n{nota+1}"]
    media = soma/4

    # Verificação da situação
    if media >= 7:
        situacao = "Aprovado"
    else:
        situacao = "Reprovado"

    # Imprime o aluno, notas e situação
    texto  = f"{aluno['nome']:<20} "
    texto += f"{aluno['n1']:<6.1f} "
    texto += f"{aluno['n2']:<6.1f} "
    texto += f"{aluno['n3']:<6.1f} "
    texto += f"{aluno['n4']:<6.1f} "
    texto += f"{media:<10.1f} "
    texto += f"{situacao:<5}"

    print(texto)
```
