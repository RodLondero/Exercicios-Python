# Exercício 04

Leia e armazene em um dicionário o nome, a idade e o número do telefone de seus 
contatos, sendo que a chave deve ser o nome. Ao digitar uma string vazia para o nome, o programa interrompe a leitura e se encerra. 

Apresente na tela os dados lidos em ordem alfabética pelo nome dos contatos. Uma possível solução de ordenar alfabeticamente é usar o método sort. 

Em seguida, armazene os contatos em outros dois dicionários, utilizando como 
critério a idade: menores de 18 anos em um e os maiores em outro dicionário, eliminando o original. Apresente na tela os dois dicionários resultantes da separação.

Imprima na tela um teste do seu programa usando como primeiro cadastro o seu 
nome, como telefone o seu RU, e como idade os dois últimos dígitos do seu RU


```python
# Inicialização das variáveis
contatos = []
contato = {}
menorIdade = []
maiorIdade = []

# Loop para receber os dados de entrada
while True:
    print("\n")

    # Solicita o nome ao usuário e repete até digite um nome válido
    # - Não pode ser um nome com apenas números
    contato["nome"] = input("Nome: ")

    while contato["nome"].isnumeric():
        print("\nNome não pode ser um número.")
        contato["nome"] = str(input("Nome: "))

    # Se for digitado um nome em branco, interrompe o Loop
    if contato["nome"].strip() == "":
        break

    # Solicita a idade ao usuário e repete até digite uma idade válida
    # - Apenas números
    while True:
        try:
            contato["idade"] = int(input("Idade: "))
            break
        except Exception:
            print("\nDigite apenas números para a idade.")

    # Solicita o telefone ao usuário e repete até digite uma telefone válido
    # - Apenas números
    while True:
        try:
            contato["fone"] = int(input("Telefone: "))
            break
        except Exception:
            print("\nDigite apenas números para o telefone.")

    # Guarda o contato em uma lista
    contatos.append(contato.copy())

# Ordena a lista dos contatos pelo nome
contatos = sorted(contatos, key=lambda k: k["nome"])

# Imprime os contatos ordenados pelo nome
print("-" * 52)
print("Contatos ordenados pelo nome")
print("-" * 52)
print(f'{"Nome":<30}{"Idade":<10}{"Telefone":<15}')
for contato in contatos:
    print(f"{contato['nome']:<30}{contato['idade']:<10}{contato['fone']:<15}")

# Separa os contatos entre maior e menor idade (18 anos)
for contato in contatos:
    if contato["idade"] < 18:
        menorIdade.append(contato.copy())
    else:
        maiorIdade.append(contato.copy())

# Limpa a lista original de contatos
contatos.clear()

# Imprime os contatos com 18 anos ou mais
print("\n" + "-" * 52)
print("Maiores de idade")
print("-" * 52)
print(f'{"Nome":<30}{"Idade":<10}{"Telefone":<15}')
for contato in maiorIdade:
    print(f"{contato['nome']:<30}{contato['idade']:<10}{contato['fone']:<15}")

# Imprime os contatos com menos de 18 anos
print("-" * 52)
print("Menores de idade")
print("-" * 52)
print(f'{"Nome":<30}{"Idade":<10}{"Telefone":<15}')
for contato in menorIdade:
    print(f"{contato['nome']:<30}{contato['idade']:<10}{contato['fone']:<15}")

# Imprime os valores das listas maiorIdade e menorIdade
print()
print("-" * 52)
print("Maiores de idade", maiorIdade)
print("Menores de idade", menorIdade)
```
